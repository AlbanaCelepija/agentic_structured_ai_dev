```python
import requests
import json
```

```python

def ask_ollama(prompt):
    url = "http://localhost:11434/api/generate"
    data = {
        "model": "llama3.2",
        "prompt": prompt,
        "stream": False
    }
    response = requests.post(url, json=data)
    return response.json()["response"]
```

```python

plan = """
You are an intelligent agent. Here is the plan:

1. Summarize the following text: "The European Commission released a new strategy for AI..."
2. Extract keywords.
3. Suggest one potential research question.

Execute the plan step by step. Do NOT skip steps.
"""

result = ask_ollama(plan)
print(result)
```
